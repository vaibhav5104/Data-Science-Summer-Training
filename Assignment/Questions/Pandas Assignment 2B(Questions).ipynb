```python
data.columns = data.columns.str.strip()
```

```python
# Combine the first three columns into a datetime index
data['datetime'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].agg('-'.join, axis=1), format='%y-%m-%d')

# Set the datetime column as the index
data.set_index('datetime', inplace=True)

# Drop the original Yr, Mo, Dy columns
data.drop(columns=['Yr', 'Mo', 'Dy'], inplace=True)
```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
def fix_year(year):
    # Adjust the year to be 2000s if it's in the 1900s
    if year < 100:
        return 2000 + year
    return year
# Apply the function to the 'Yr' column
data['Yr'] = data['Yr'].apply(fix_year)

# Create a 'datetime' column
data['datetime'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1), format='%Y-%m-%d')

print(data)
```
